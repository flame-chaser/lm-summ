```python
'''
НЕ СТИРАТЬ, ПАРСЕР ФИН ГПТ
'''
from time import sleep
from tqdm.notebook import tqdm
for i in tqdm(range(df.shape[0])):
    browser.find_element(By.ID, 'i-name').clear()
    browser.find_element(By.ID, 'i-name').send_keys(str(df.loc[i, 'Полное наименование организации']))
    browser.find_element(By.ID, 'i-content').clear()
    browser.find_element(By.ID, 'i-content').send_keys(str(df.loc[i, 'Данные']))
    sleep(1)
    browser.find_element(By.NAME, 'submit').click()
    res = WebDriverWait(browser, 120).until(EC.presence_of_element_located((By.ID, 'ai-results')))
    field = WebDriverWait(res, 120).until(EC.presence_of_element_located((By.CLASS_NAME, 'card-body')))
    df.loc[i,'Отчёт'] = field.text
    print(df.loc[i])
```

```python
'''
НЕ СТИРАТЬ, ПАРСЕР ФИН ОТЧЁТОВ
'''
from time import sleep
from tqdm.notebook import tqdm
flag=False
for i in tqdm(range(260,df.shape[0])):
    browser.find_element(By.ID, 'search').clear()
    browser.find_element(By.ID, 'search').send_keys(str(df.loc[i, 'Идентификационный номер налогоплательщика'])+Keys.ENTER)
    sleep(2)
    elems = browser.find_elements(By.CLASS_NAME, 'results-search-table-row')
    if len(elems) > 0:
        elems[-1].click()
    else:
        continue
    sleep(2)
    if not flag:
        browser.find_elements(By.CLASS_NAME, 'download-reports-wrapper')[-1].click()
        flag=True
        sleep(2)
    browser.find_elements(By.CLASS_NAME, 'form-item-checkbox_checkbox')[2].click()
    sleep(1)
    wrapper = browser.find_element(By.CLASS_NAME, 'download-reports-footer')
    btn = wrapper.find_element(By.CLASS_NAME, 'button_md')
    coordinates = btn.location_once_scrolled_into_view # returns dict of X, Y coordinates
    browser.execute_script('window.scrollTo({}, {});'.format(coordinates['x'], coordinates['y']))
    sleep(1)
    btn.click()
```
